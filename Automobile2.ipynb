```python
from sklearn.preprocessing import LabelEncoder

label=LabelEncoder()
df['fuel']=label.fit_transform(df.fuel)
df['owner']=label.fit_transform(df.owner)
df['seats']=label.fit_transform(df.seats)
df['Brand']=label.fit_transform(df.Brand)
df['transmission']=label.fit_transform(df.transmission)
```

```python
df['mileagev'] = pd.to_numeric(df['mileagev'], errors='coerce').astype('Float32')
df['enginev'] = pd.to_numeric(df['enginev'], errors='coerce').astype('Int64')
df['max_powerv'] = pd.to_numeric(df['max_powerv'], errors='coerce').astype('Float32')
```

```python
def Brand(brand_name):
  df2=df[(df['Brand']==brand_name)]
  return df2

brand_name=input("Input the brand name for model comparision: ")
res=Brand(brand_name)
figure=px.bar(res.sort_values('mileagev', ascending=False).head(), x='name', y='mileagev',color='selling_price',template="simple_white",
              title='--------')
figure.show()
```

```python
from sklearn.model_selection import train_test_split
y = df['selling_price']
X = df.drop(['name', 'year', 'fuel', 'seller_type','seats',
       'torquev', 'model_name'], axis=1)
x_train, x_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
from sklearn.ensemble import RandomForestRegressor
model = RandomForestRegressor(n_estimators=500, random_state=1)
model.fit(x_train, y_train)
prediction = model.predict(x_test)
```

```python
from sklearn.metrics import mean_absolute_error
print(mean_absolute_error(y_test,prediction))
```

```python
from sklearn.metrics import r2_score
print(r2_score(y_test,prediction))
```
